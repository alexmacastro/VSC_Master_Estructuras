Consideremos una membrana rectangular, definida como sigue: 

* Dominio rectangular, $[0,a] \times [0,b]$
* Espesor de la membrana $h$.
* Tensión de la membrana $T$.
* Densidad del material $\rho$.

La siguiente figura muestra la configuración de esta membrana.

```python
from IPython.display import Image

Image(filename='membrana_rectangular.png')
```

Los modos de vibración de esta membrana tienen la siguiente expresión: 

\begin{equation}
\phi_{mn}(x,y) = \mathrm{sen} \left(\dfrac{m\,\pi\,x}{a}\right) \cdot \mathrm{sen} \left( \dfrac{n\,\pi\,y}{b} \right)
\end{equation}

Las frecuencias naturales (angulares) $\omega_{mn}$ tienen la siguiente expresión: 

\begin{equation}
\omega_{mn} = \pi\,\sqrt{\dfrac{T}{\rho\,h}\,\left[ \left(\dfrac{m}{a}\right)^2 + \left( \dfrac{n}{b}\right)^2\right]}
\end{equation}

Para una membrana vibrando en un modo de vibración, tendrá la siguiente expresión: 

\begin{equation}
w_{mn}(x,y,t) = \phi_{mn} (x,y) \,\left[ A_{mn} \,\mathrm{sen}(\omega_{mn}\,t) + B_{mn} \, \cos(\omega_{mn}) \,t  \right]
\end{equation}

En lo que sigue, representaremos la solución para los diferentes modos de vibración. 


```python
from __future__ import division
%pylab inline 

#==============================================================================
# Datos de entrada
#==============================================================================

A = 1    # Lado a, dimensión (m)
B = 1    # Lado b, dimensión (m)
H = 0.01 # Espesor de la membrana (m)

Tens = 10    # Tensión de membrana, dato en N / m 
Rho = 100      # Densidad VOLUMETRICA, dato en kg/m3
```

## Funciones

```python
def Omegamn(m, n):
    Fmn = (m / A)**2. + (n/B)**2.
    wmn = np.pi * np.sqrt( Tens / (Rho * H) * Fmn)
    return wmn 


print "Frecuencia Angular Fundamental ", Omegamn(1,1) , " (rd/s)" 

print "Frecuencia (Hz) Fundamental", Omegamn(1,1) / ( 2. * np.pi ), " (Hz)"

def Modo(m, n, x, y):
    return np.sin (m * np.pi * x / A) * np.sin (n * np.pi * y / B)
```

```python
Mmodo = 1
Nmodo = 1

T = 2.*np.pi / Omegamn(Mmodo,Nmodo) 
t = np.arange(0, 2 * T, 2* T / 100) # Vector de tiempos.
len(t)

from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(0, A, 0.02*A)
Y = np.arange(0, B, 0.02*B)
X, Y = np.meshgrid(X, Y)
Z = Modo(Mmodo, Nmodo, X, Y)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(-1.01, 1.01)
ax.set_xlim(0,A)
ax.set_ylim(0,B)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

# Figuras Chladni (Vibración en Placas)

Las formas modales pueden visualizarse por vía experimental, mediante el método de [Chladni](https://en.wikipedia.org/wiki/Ernst_Chladni). 

La técnica consiste en introducir un conjunto de partículas pequeñas, sobre una placa, y generar una excitación a una frecuencia simple. Las particulas se desplazan hacia las posiciones de las líneas nodos (puntos sin desplazamiento), permitiendo así visualizar el modo de vibración.
